# Sentiment trends visualization

```python
df.info(verbose=True)
```

```python
df.head(2)
```

```python
df_by_week = df.groupby(pd.Grouper(key='CREATED_AT', freq='W'))
```

```python
weekly_neg_score = df_by_week["NEGATIVE"].mean()
weekly_pos_score = df_by_week["POSITIVE"].mean()
# weekly_neg_score = weekly_neg_score.loc['2021-01-01': '2022-01-01']
# weekly_pos_score = weekly_pos_score.loc['2021-01-01': '2022-01-01']
```

```python
sentiment_trend = pd.merge(weekly_neg_score, weekly_pos_score, left_index=True, right_index=True)
sentiment_trend.plot(title = "Zendesk ticket sentiment trend", figsize = (25,5))
```
